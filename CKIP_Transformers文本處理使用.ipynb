## 使用 CKIP-Transformers

Github:https://github.com/ckiplab/ckip-transformers

```python
!pip install transformers
import transformers
!pip install ckip_transformers
import ckip_transformers

from ckip_transformers.nlp import CkipWordSegmenter
from ckip_transformers.nlp import CkipPosTagger
from ckip_transformers.nlp import CkipNerChunker
```

```python
# 載入CKIP model
# device = -1 代表用CPU、device = 0 代表用GPU

print("Initializing drivers ... WS")
CKIP_WS = CkipWordSegmenter(model = "bert-base",device = 0)  #建立斷詞器， device = 0 使用GPU

print("Initializing drivers ... POS")
CKIP_POS = CkipPosTagger(model = "bert-base",device = 0)   #建立詞性標註器

print("Initializing drivers ... NER")
CKIP_NER = CkipNerChunker(model = "bert-base", device = 0)  ##建立NER標註器
```

```python
# 詞性標記工具會自動，。!"? 等字元在執行模型前切割句子(輸出的句子會自動接回)
#可設定delim_set數使用別的字元做切割·
#另外可指定 use_delim = False 已停用此功能或於斷詞、實體辨識時指定 use_delim = False已啟用此功能·

# Disable sentence segmentation
# CKIP_POS(text, use_delim = Talse)
# Use new line characters and tabs for sentence segmentation
# CKIP_POS(ws, delim_set = "\n\t")
```

## 斷詞

詞性標記工具會自動用 ，。!"?等字元在執行模型前切割句子(輸出的句子會自動接回)
可設定delim_set數使用別的字元做切割·

另外可指定 use_delim = False 已停用此功能或於斷詞、實體辨識時指定 use_delim = False已啟用此功能·

```python
zh_text = ['人工智慧亦稱智械、機器智慧,指由人製造出來的機器所表現出来的智慧。',
'通常人工智慧是指透過普通電腦程式來现人题智慧的技術。',
'中華郵政未来智慧物流服務,將取之大眾智慧,中華郵政帶給民眾更好的便利生活。']
```

```python
#進行斷詞
#斷詞的輸入必須是 list of sentences  (一次餵全部)
#也就是多個字串的方式輸入
C_sents_annotated_ws = CKIP_WS(zh_text)

print("\nC_sents_annotated_ws:",C_sents_annotated_ws)
```

## 詞性標註(POS)

```python
# 詞性標註的輸入必須是list of list of words
C_sents_annotated_pos = CKIP_POS(C_sents_annotated_ws)  #放入的是前面已斷過的句子

print("\n C_sents_annotated_pos:",C_sents_annotated_pos)
print(list(zip(C_sents_annotated_ws[0],C_sents_annotated_pos[0])))
```

```python
C_sents_annotated_pos
```

### 根據詞性篩選字詞

```python
# 篩選命名實體
def filter_by_pos(sent,pos_filter):
  filter_pos_sent = []
  for j,(word,pos) in enumerate(zip(sent[0],sent[1])):
    if pos[0] in pos_filter: #把每個字的開頭拿來做判斷
      filter_pos_sent.append(word)
  return filter_pos_sent
```

```python
# 定義要篩選的詞性 tag
pos_filter = ['N']   #若是名詞

# 合併斷詞和詞性
C_sents_annotated_ws_pos = zip(C_sents_annotated_ws, C_sents_annotated_pos)

# 根據特定詞性篩選
C_sents_filter_pos = [filter_by_pos(sent, pos_filter) for sent in C_sents_annotated_ws_pos]
```

```python
print("sents_annotated_ws: ",C_sents_annotated_ws,'\n')
print("sents_annotated_pos: ",C_sents_annotated_pos,'\n')
print("sents_filter_pos: ",C_sents_filter_pos,'\n')
```

## 命名實體辨識(NER)

```python
## NER的輸入必須是list of sentences
C_sents_annotated_ner = CKIP_NER(zh_text)
```

```python
print(C_sents_annotated_ner)   #第一、二個為[]:代表句子中沒有找到NER，第三句才有，而且有兩個
```

### 從NER中將名詞獨立出來

```python
def filter_by_ner(sent, ner_filter):
  filter_ner_sent = []
  for j,word in enumerate(sent):
    if word.ner in ner_filter:
      filter_ner_sent.append(word.word)
  return filter_ner_sent
```

```python
# 定義要篩選的命名實體tag
ner_filter = ['ORG']

# 根據特定詞性篩選
C_sents_filter_ner = [filter_by_ner(sent, ner_filter) for sent in C_sents_annotated_ner]

print("Sents_filter_ner", C_sents_filter_ner)
```
